Problema 8 de derivación -- Es posible construir una aproximaci ́on de orden $$O(h^2)$$ para la derivada progresiva. Para tal proposito, se escribe el polinomio de interpolaci ́on de grado 2 para el conjunto soporte $$Ω = {(x_0, f(x_0)), (x_1, f(x_1)), (x_2, f(x_2))}$$, y posteriormente se calcula la derivada de este polinomio.

a) Calcular analıticamente el polinomio que interpola el conjunto soporte.


b) Derivar el polinomio interpolador para encontrar la derivada en el punto $$x_0$$:
$$f′(x_0)≈ p′(x_0) = \frac{1}{2h}(−3f(x_0)+4f(x_1)−f(x_2))$$. Si la descritización es equidistante, entonces: $$f′(x) = \frac{1}{2h}(−3f(x)+4f(x+h)−f(x+2h))$$

c) (Python) Para $$f(x) = \sqrt{tan(x)}$$ estimar la derivada progresiva de orden $$O(h^2)$$ (expresión anterior) en el intervalo [0.1, 1.1] con h = 0.01.

```python
# Definir la función
def f(x):
    return np.sqrt(np.tan(x))

# Definir la derivada progresiva de orden O(h^2)
def derivada_progresiva(f, x, h):
    return (-3*f(x) + 4*f(x + h) - f(x + 2*h)) / (2*h)

# Intervalo y paso
x = np.arange(0.1, 1.1, 0.01)
h = 0.01

# Calcular la derivada progresiva en cada punto del intervalo
derivadas = [derivada_progresiva(f, xi, h) for xi in x]

# Graficar la función y su derivada
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label='f(x) = sqrt(tan(x))')
plt.plot(x, derivadas, label="f'(x) (aproximación progresiva O(h^2))", linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("Aproximación de la derivada progresiva de f(x) = sqrt(tan(x))")
plt.grid(True)
plt.show()
```

d) (Python) Para $$f(x) = \sqrt{tan(x)}$$ estimar la derivada central de orden $$O(h^2)$$ en el intervalo [0.1, 1.1] con h = 0.01.

```python
# Definir la derivada central de orden O(h^2)
def derivada_central(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

# Intervalo y paso
x = np.arange(0.1, 1.1, 0.01)
h = 0.01

# Calcular la derivada central en cada punto del intervalo
derivadas = [derivada_central(f, xi, h) for xi in x]

# Graficar la función y su derivada
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label='f(x) = sqrt(tan(x))')
plt.plot(x, derivadas, label="f'(x) (aproximación central O(h^2))", linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("Aproximación de la derivada central de f(x) = sqrt(tan(x))")
plt.grid(True)
plt.show()
```
