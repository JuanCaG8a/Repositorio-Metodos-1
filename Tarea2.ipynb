c) (Python) Para $$f(x) = \sqrt{tan(x)}$$ estimar la derivada progresiva de orden $$O(h^2)$$ (expresión anterior) en el intervalo [0.1, 1.1] con h = 0.01.

```python
# Definir la función
def f(x):
    return np.sqrt(np.tan(x))

# Definir la derivada progresiva de orden O(h^2)
def derivada_progresiva(f, x, h):
    return (-3*f(x) + 4*f(x + h) - f(x + 2*h)) / (2*h)

# Intervalo y paso
x = np.arange(0.1, 1.1, 0.01)
h = 0.01

# Calcular la derivada progresiva en cada punto del intervalo
derivadas = [derivada_progresiva(f, xi, h) for xi in x]

# Graficar la función y su derivada
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label='f(x) = sqrt(tan(x))')
plt.plot(x, derivadas, label="f'(x) (aproximación progresiva O(h^2))", linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("Aproximación de la derivada progresiva de f(x) = sqrt(tan(x))")
plt.grid(True)
plt.show()
```

d) (Python) Para $$f(x) = \sqrt{tan(x)}$$ estimar la derivada central de orden $$O(h^2)$$ en el intervalo [0.1, 1.1] con h = 0.01.

```python
# Definir la derivada central de orden O(h^2)
def derivada_central(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

# Intervalo y paso
x = np.arange(0.1, 1.1, 0.01)
h = 0.01

# Calcular la derivada central en cada punto del intervalo
derivadas = [derivada_central(f, xi, h) for xi in x]

# Graficar la función y su derivada
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label='f(x) = sqrt(tan(x))')
plt.plot(x, derivadas, label="f'(x) (aproximación central O(h^2))", linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("Aproximación de la derivada central de f(x) = sqrt(tan(x))")
plt.grid(True)
plt.show()
```

e) Calcule analiticamente la derivada de la funcion f(x), y grafique con la estimacion central y progresiva de orden $$O(h^2)$$.

```python
def f(x):
    return np.sqrt(np.tan(x))

def f_prime(x):
    return (1 / (2 * np.sqrt(np.tan(x)))) * (1 / np.cos(x)**2)

# Aproximaciones numéricas
def forward_diff(f, x, h):
    return (f(x + h) - f(x)) / h

def central_diff(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

# Valores de x y h
x = np.linspace(0.1, 1.4, 100)
h = 0.01

# Derivadas
f_prime_exact = f_prime(x)
f_prime_forward = forward_diff(f, x, h)
f_prime_central = central_diff(f, x, h)

# Errores
error_forward = np.abs(f_prime_exact - f_prime_forward)
error_central = np.abs(f_prime_exact - f_prime_central)

# Gráficas
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, f_prime_exact, label='Exacta')
plt.plot(x, f_prime_forward, label='Progresiva')
plt.plot(x, f_prime_central, label='Centrada')
plt.legend()
plt.title('Derivadas')

plt.subplot(1, 2, 2)
plt.plot(x, error_forward, label='Error Progresiva')
plt.plot(x, error_central, label='Error Centrada')
plt.legend()
plt.title('Errores')

plt.show()
```
